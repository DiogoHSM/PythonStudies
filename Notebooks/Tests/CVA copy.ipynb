```python
import matplotlib.pyplot as plt

points = {
    'A': (0, 0),
    'B': (0, 1),
    'C': (1, 1.5),
    'D': (2, 1),
    'E': (1.5, 0)
}

x = [point[0] for point in points.values()]
y = [point[1] for point in points.values()]

plt.scatter(x, y)
for label, point in points.items():
    plt.annotate(f'{label} ({point[0]}, {point[1]})', point, textcoords="offset points", xytext=(0,10), ha='center')

plt.axis('off')
plt.show()
```
