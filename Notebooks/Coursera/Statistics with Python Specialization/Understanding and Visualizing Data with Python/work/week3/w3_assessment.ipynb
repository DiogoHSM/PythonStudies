## Scatter plots

First, let's looks at two variables that we expect to have a strong relationship, 'Height' and 'Wingspan'.

````python
# Make a Seaborn scatter plot with x = height and y = wingspan using sns.scatterplot(x, y)
# Create the scatter plot
sns.scatterplot(x=df['Height'], y=df['Wingspan'], hue=df['Gender'])

# Calculate the correlation matrix for each gender
correlation_matrix = df.groupby('Gender')[['Height', 'Wingspan']].corr()

# Calculate the interquartile range for height and wingspan
iqr_height = np.percentile(df['Height'], 75) - np.percentile(df['Height'], 25)
iqr_wingspan = np.percentile(df['Wingspan'], 75) - np.percentile(df['Wingspan'], 25)

# Print the correlation matrix and interquartile range
print("Correlation Matrix:")
print(correlation_matrix)
print("Interquartile Range for Height:", iqr_height)
print("Interquartile Range for Wingspan:", iqr_wingspan)
````
